```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

checkpoint = "Deci/DeciLM-6b-instruct"
device = "cpu" # for GPU usage or "cpu" for CPU usage

tokenizer = AutoTokenizer.from_pretrained(checkpoint, cache_dir=".cache")
model = AutoModelForCausalLM.from_pretrained(checkpoint, torch_dtype=torch.bfloat16, trust_remote_code=True, cache_dir=".cache").to(device)
```

```python
def generate(prompt):
    inputs = tokenizer.encode(prompt, return_tensors="pt").to(device)
    outputs = model.generate(inputs, max_new_tokens=16, do_sample=True, top_p=0.95, pad_token_id=tokenizer.eos_token_id)
    return tokenizer.decode(outputs[0])
    
prompt = input("Enter a prompt: ")
print(prompt, end="")

while True:
    out = generate(prompt)
    print(out, end="")
    prompt += out
```
